# `Where Does Anaconda install on a Mac?`

If you used the normal installer from the webpage (GUI) you will find it inside of

`/opt/anaconda3` 

If you used the curl request or NON-GUI installation you should see it in here:

`/Users/YourUserName/anaconda3`

# `Base Environment`

Default storage location for Anaconda when you open a terminal is `(base)`. Think of this as a virtual environment separate from your system wide Python for example. Here you do not have to worry about other installations of Python or packages having issues or conflicts because it is self contained. `BUT,DO NOT GO ABOUT INDIVIDUALLY UPDATE SEPARATE LIBRARIES FOR PROJECTS HERE`. This is the area where your application is stored and each environment you create to work in will be little files/folders containing your work with whatever dependencies and packages needed without affecting global installations. 
+ The `(base)` environment should not be used to store your dependencies or furth update everything as you work on projects.
    + Use this default as a place to update everything when newer versions of Anaconda are available and then separate environments for projects and version control
 
A better explanation by offical documentation is here: 

https://www.anaconda.com/docs/tools/working-with-conda/environments

# Commands

+ `Create Environments`

**`conda create  - - name  yourName_of_env `**

+ `Create Environments & Add Packages with specific version`

**conda create  - - name  yourName_of_env python=3.8**


+ `Activating/deactivating environments`

**`conda activate EnvironmentName`**   or  **`conda deactivate EnvironmentName`**

+ `Checking list of packages we have in our environment`

**conda list**

+ `Check the environments we have`
	
    **conda env list**


+ `View List of Current Environments`

**conda info --envs**

+ `Update Package/Libraries`

**conda update _packageName**

+ Installing packages

**conda install _andYourPackageName_**

    + Installing NON-Anaconda packages will need to use *conda-forge* or *pip* for example!

+ `Uninstalling a package/library`

	**conda remove  TypeYourPackageName**

# `Link Jupyter Notebook with Terminal ENV's:`

When you are using the terminal you can create, delete and work with as many virtual environments as you need. But, making this interect with the `Jupyter Notebook` you need to setup everything so that the *kernels* are able to read in the environments.

+ You will need to Install `nb_conda_kernels`: put this in the virtual environment that you will be storing Jupyter Notebook such as *(base)*
    + This will allow (base) for example to interact with Jupyter allowing you to create separate environments without needing to install every dependency for Jupyter which is cumbersome. Instead you create a link between each environment and (base) which is like a pointer.

+ In each of the environments that you create you will add: `ipykernels`


1) Add Environment to Anaconda if you have not already. This will be used for version control as well as sharing and storing your work.

**conda create -n env_name_youWant**

2.) we need to *activate* the environment we created with the name "env_name_youWant"

**conda activate env_name_youWant**

3.) Install *ipykernel* into this environment. This will allow us to use python in our environment within Jupyter Notebook. There are different kernels for varying languages such as R-Studio.

**conda install ipykernel**

4.) Deactivate the current environment because we need to globally set a link between our terminal environments and Jupyter Notebook GUI

**conda deactivate**

5.) The *(base)* environment which is usually the route we go to create this link *but NOT neccessarily* the place everyone uses. Depending how your terminal is setup using *conda deactivate* puts you in a *(base)* environment. If not use command:

**conda activate base**

6.) The link/"pointer" we are going to use will be `nb_conda_kernels` allowing each session of a new environment we create to connect with Jupyter Notebook or Jupyter Lab. This may be your *(base)* environment but it is not necessary. There are circumstances when you may need/forced to change this. If we install *nb_conda_kernels* we are allowing any notebook run from this environment to automatically show kernels from any other environment that has *ipykernels* installed.

**conda install nb_conda_kernels**

or 

*conda install -n YourEnvName nb_conda_kernels*


`Here is an important distinction and clarification:` the above install of `nb_conda_kernels` will create an automatic creation of your environment to be found within Jupyter as long as you are calling jupyter from the location of installation of `nb_conda_kernels` and your other environments have the `ipykernel or similar` installed. But, you may have cases when this is not required to have everything showup in notebooks. It can become cumbersome.

An alternative is to do this case by case:

****



7.) We can check the kernels available for Jupyter using:

**jupyter kernelspec list**


8.) You are ready to connect to Jupyter Notebook (may need to restart kernel) now and look for your new environment that was created in the terminal to showup in the Jupyter Notebook drop-down for kernels. 

+ After you install the *ipykernel* in your environment
+ 
+ run the commmand:

**ipython kernel install --user --name=new-env**

or

**python -m ipykernel install --user --name "$YourEnvName"**

`If you want the displayed name not to be same as ENV name in Jupyter do this command:`

**python -m ipykernel install --user --name "$YourEnvName" --display-name "Python [YourCustomNameHere"]"**


9.) If you want to *unregister* the environment from the kernel, meaning that you do not want it to connect to Jupyter Notebook:

**jupyter kernelspec uninstall YourEnvName**

*kernels* are needed to interact with each language for example python has `ipykernel`, r-studio: `r-irkernel`, etc.

# Uninstall Anaconda from Mac:

Uninstall Anaconda:

`rm -rf anaconda3`

`rm -rf ~/anaconda3`

`sudo rm -rf /opt/anaconda3`
    
There are additional hidden files to consider removing:
	
`rm ~/.condarc ~/.conda ~/.continuum`

https://www.anaconda.com/docs/getting-started/anaconda/uninstall
