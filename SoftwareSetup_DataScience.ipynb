# `Trouble Shooting MongoSH Install:`

+ If you get a message saying that `zsh: command not found: mongosh` you will need to troubleshoot

1.) recheck the file path of `mongosh file`
    
2.) recheck `.zshrc` file is correct formatting or location

3.) go into the file path of where you stored `mongosh-2.5.0-darwin-arm64` this file and it matches `.zshrc` file

4.) check the contents of folder: `Users/YourUserName/mongosh-2.5.0-darwin-arm64/bin` there should be a `mongosh` file

5.) make sure there is a connection to `mongod` and nothing is interfering

6.) remove file and start again with `rm -rf Users/YourUserName/mongosh-2.5.0-darwin-arm64` and redo everything

7.) check if there are any issues in the `settings -> system preferences` of device blocking you.

8.) consider `sudo` and your password to setup everything
