# `MongoDB:`

We need to go to the MongoDB website and choose the `community edition` if we are not payinf for the service or involved with a business account.

**`Tar File Unzipped:`**

https://www.mongodb.com/docs/manual/tutorial/install-mongodb-on-os-x-tarball/

The installation steps are outlined here but may seem confusing.

1.) **Download the file and unzip if needed**

+ `tar -zxvf file_name_folder`
+ Now, in terminal `cd /Downloads`

2.) **Manage Binaries:** 

+ We need to take this folder of MongoDB that was downloaded and send it to the `home directory` we are working with.
    + If you are in the terminal by default: `pwd` and this will print the home directory for you, it may have a tilde i nthe prompt name before the command.
+ Then we will make a folder to store this download unless you want to move directly

`sudo mv yourMongoFileName locationHomeFoler` -> `sudo namefile /User/yourUserName`

or 

the default area mac would store 3rd party apps: `sudo mv yourMongoFolder /usr/local` and if you want you can add a folder prefixed with mongodb as a reference like this:

+ `cd /usr/local`
+ `mkdir mongodb`
+ then `sudo mv yourMongoFolder /usr/local/mongodb`

+ If you do not have the bin folder later, when you encounter: `/usr/local/bin` you will need to create or delete what looks like a folder buy may be a file. More on this later.

3.) Set the path for MongoDB binaries to allow `mongod` command to be in PATH:

This will be found in your bash or zsh file. I am using Zsh, so the variable we need to bind will be in the `.zshrc` file.

Using Vim in the terminal: 

+ you can find the `.zshrc` file if we are in the home folder of terminal `Users/yourusername` is what I am describing
+ use command: `ls -al` to look for this file `.zshrc` if it exists good.
    + If not create it using `touch .zshrc` and save it.
+ Then you will need to use this file to store the variables to connect our binaries
    + open file with your text editor, i am using vim: `vim .zshrc`
    + inside vim press *i* for insert and type this: `export PATH="/usr/local/mongodb/bin:$PATH"` or `export PATH="/User/yourusernamehere/mongodb/bin:$PATH"`
    + to exit vim: `:wq` to save and quit
    + then to acitivate your new setting: `source .zshrc` or restart terminal and run command `mongod`
        + You will see it trying to work but needs a connection to the db path that was not setup yet.


3.) **Create Data/db Folder(s)** you can do this in 1 step instead of 2. 

+ This is used to *let MongoDB know where to write data*.

`sudo mkdir -p data/db`

Depending where you want to install this be in correct and consistent location in terminal: `Users/yourusername` or `/usr/local`

4.) We need the data to be in the path of mongod now: make sure you are in the path of where MONGO is stored during setup
such as home directory: `Users/youruserName` or `/usr/local` before you run this command to bind them.

`sudo mongod --dbpath=/Users/yourusername/data/db` or `sudo mongod --dbpath=/usr/local/yourmongoDBNameHere/data/db`


5.) **Create Log Folder** to store/redirect our log reports, which can be helpful for troubleshooting

`sudo mkdir -p logs`

6.) **Next, where to send log reports:** 

`./mongod --dbpath /usr/local/mongodb/your/data/db > /usr/local/mongodb/mongodb-macos-x86_64-5.0.14/logs/mongod.log 2>&1 &`

or

`./mongod --dbpath /Users/yourUserName/data/db > /Users/yourUserName/mongodb/mongodb-macos-x86_64-5.0.14/logs/mongod.log 2>&1 &`


And change this to your version: *mongodb-macos-x86_64-5.0.14*


8.)

Next, you will need the `Mongosh` or `GUI` this will be used to connect the `server` running `MongoDB` from the command `mongod` that was used to initiate the server. But, you need a way to interact with the server/client. The old way was to use `mongo` but this was depricated and `mongosh` is used now. In order to use this command we need to do another download and create a path for the binaries again.

https://www.mongodb.com/docs/mongodb-shell/install/



9.)

+ After you get Mongod started you can use the shell or similar and check its location: `db.serverCmdLineOpts()`

# `This is a small rant but useful:`

You will see people doing copy pasta for code or tutorials but they don't get always understand. I fall into this sometimes as well. So when Mac/Apple products usually download third-party applications they will go in the `usr/local` folder as a place to store data. 

+ Now, `Anaconda` is a package manager but it stores files differently in its own container/storage location.
    + Run the command: `conda info` and you will see detailed install information and files
+ `Homebrew` will store files in a `/usr/local/Cellar` location for example.

When I was watching tutorials for this installation of `MongoDB` I noticed 1 video featuring the storage location in the `Home directory: Users/YourUserName` but I then saw that the official site says `/usr/local` which had me confused. Now you can store the file here as long as you MAKE SURE, your symbolic links match for dependencies/binaries. But, understand that you need to install and have a general idea where folders are stored and how to retreive/communicate to them ok!

If you have issues with `/usr/local/bin` try to:
+ determine if it is there or not
    + `cd /usr/local`
        + `ls`
    + check for bin folder
        + if it is not there you will need to create directory.
        + if it is there go to next step
          
+ check if it is there and is it a file or directory
`file /usr/local/bin`


the result will say if it is a file or a directory

+ if it is a file, delete first then make a directory
   + `rm bin`
   + make folder bin: `sudo mkdir -p /usr/local/bin`
   + you will now be asked for your admin password
   + Get permissions: `sudo chown -R $(id -nu):admin /user/local/bin
